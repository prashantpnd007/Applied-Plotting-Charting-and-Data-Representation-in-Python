```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)
n=20


import matplotlib.animation as animation

# Writer = animation.writers['imagemagick']
# writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
# Writer = animation.writers['ffmpeg']
# writer = Writer(fps=50)

def update(curr):
    from random import randrange
    pq = randrange(100, 1000)
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == n: 
        a.event_source.stop()
    
    bins = np.arange(-4, 4, 0.5)
    plt.subplot(2, 2, 1)
    plt.cla()
    plt.hist(x1[:pq], bins=40)
    plt.axis([-4,1,0,300])
    plt.gca().set_title('Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [0,250])
    
    plt.subplot(2, 2, 2)
    plt.cla()
    plt.hist(x2[:pq], bins=40)
    plt.axis([-1,5,0,300])
    plt.gca().set_title('Gamma Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [4,250])
    
    plt.subplot(2, 2, 3)
    plt.cla()
    plt.hist(x3[:pq], bins=20)
    plt.axis([6,30,0,400])
    plt.gca().set_title('Exp Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [25,350])
    
    plt.subplot(2, 2, 4)
    plt.cla()
    plt.hist(x4[:pq], bins=20)
    plt.axis([6,30,0,100])
    plt.gca().set_title('Uniform Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [25,80])
    
fig = plt.figure(figsize=(10,10))
a = animation.FuncAnimation(fig, update, interval=800)    
# a.save('/tmp/animation.gif', writer='imagemagick', fps=30)
a.save("./output.gif", writer='PillowWriter', fps=10, bitrate=-1)
```
