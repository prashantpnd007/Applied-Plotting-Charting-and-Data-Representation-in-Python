```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook

# reading manufacturing industry employees df
df1 = pd.read_csv('/Users/Prashant Pandey/Desktop/SMU26114603000000001A.csv')
df1.rename(columns={'SMU26114603000000001A': 'Manuf_Employees'}, inplace=True)
df1['DATE'] = pd.to_datetime(df1['DATE'])

# reading hospitality industry employees df
df2 = pd.read_csv('/Users/Prashant Pandey/Desktop/SMU26114607000000001A.csv')
df2.rename(columns={'SMU26114607000000001A': 'Hosp_Employees'}, inplace=True)
df2['DATE'] = pd.to_datetime(df1['DATE'])

# reading finance industry employees df
df3 = pd.read_csv('/Users/Prashant Pandey/Desktop/SMU26114605500000001A.csv')
df3.rename(columns={'SMU26114605500000001A': 'Fin_Employees'}, inplace=True)
df3['DATE'] = pd.to_datetime(df3['DATE'])

# reading Education industry employees df
df4 = pd.read_csv('/Users/Prashant Pandey/Desktop/SMU26114606500000001A.csv')
df4.rename(columns={'SMU26114606500000001A': 'Edu_Employees'}, inplace=True)
df4['DATE'] = pd.to_datetime(df4['DATE'])

# reading Business industry employees df
df5 = pd.read_csv('/Users/Prashant Pandey/Desktop/SMU26114606000000001A.csv')
df5.rename(columns={'SMU26114606000000001A': 'Bus_Employees'}, inplace=True)
df5['DATE'] = pd.to_datetime(df5['DATE'])

# merging all the dataframes into one
df_new1 = pd.merge(df1,df2, on='DATE', how='outer')
df_new2 = pd.merge(df_new1,df3, on="DATE", how='outer')
df_new3 = pd.merge(df_new2,df4, on="DATE", how='outer')
df_new4 = pd.merge(df_new3,df5, on="DATE", how='outer')


#converting dates in years
df_new4.rename(columns={'DATE':'Year'}, inplace=True)
df_new4['Year'] = pd.DatetimeIndex(df_new4['Year']).year

#plotting
fig, ax1=plt.subplots()

ax1.plot(df_new4['Year'], df_new4['Manuf_Employees'], color='red')
ax1.plot(df_new4['Year'], df_new4['Hosp_Employees'], color='blue')
ax1.plot(df_new4['Year'], df_new4['Fin_Employees'], color='goldenrod')
ax1.plot(df_new4['Year'], df_new4['Edu_Employees'], color='green')
ax1.plot(df_new4['Year'], df_new4['Bus_Employees'], color='black')

ax1.set_xlabel('Year')
ax1.set_ylabel('Thousands of Employees')

ax1.legend(loc='upper right', fontsize=9, frameon=True, framealpha=1)

plt.title('Employment trends in Ann Arbor, Michigan')
fig.tight_layout()
plt.show()
```
